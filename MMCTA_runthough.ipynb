#Luciferase Data-Processing Pipeline 
* Andrew Gross, Bioinformatics and Systems Biology, University of California San Diego

This work was mainly done in the Winter and Spring of 2012 in the lab of Steve Kay.  The objective was to develop a robust, automated data-processing pipelin for the lab's achive of luciferase reporter data. 

The input to the code expects folders of .TIF image stacks with titles of the form 'X-YY.TIF' where X is the run number and YY
is the two digit sequence number of the image.  This is the convention that the Kay Lab has chosen, while I have tried to make 
the pipeline robust to different systems, some sematics such as nameing may have to be manipulated to make the pipeline run. 

Should be run on 16-bit grey scale images, I have not tested on any other types, but the pipeline may still run.

It is expected that each image contains four plates.  For an example dataset see ./Data/TM249.
